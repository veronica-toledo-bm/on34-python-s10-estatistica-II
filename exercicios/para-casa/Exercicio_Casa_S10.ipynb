1. Um nutricionista quer testar se a ingest√£o di√°ria m√©dia de calorias de um grupo de 30 pessoas √© diferente de 2000 calorias.
Os dados das calorias ingeridas s√£o fornecidos abaixo. Realize o teste de hip√≥teses usando um n√≠vel de signific√¢ncia de 0,05.


```python
#lista de ingest√£o di√°ria de calorias para 30 pessoas
dados_calorias = [1900, 2100, 2050, 1980, 1950, 2100, 2000, 2150, 2200, 1850,
                  1990, 1950, 2050, 2080, 2100, 1900, 1950, 2050, 2150, 2000,
                  2200, 1900, 2100, 2000, 2150, 1850, 1990, 1950, 2050, 2080]

#Teste de hipoteses
#H0 - a media √© igual a 2000 calorias
#H1 - a media √© diferente a 2000 calorias

from scipy import stats

t_estatistico, p_valor = stats.ttest_1samp(dados_calorias, 2000)
print(f"Estatistica t √©:{t_estatistico: 1f}")
print(f"Valor p √©:{p_valor: .1f}")

if p_valor < 0.05:
  print("Rejeitamos a hipotese nula, a media √© diferente a 2000 calorias.")
else:
  print("Aceitamos a hipotese nula, porque n√£o existe efeito ou diferen√ßa.")

```

2. Um pesquisador deseja verificar se h√° uma diferen√ßa significativa entre as notas de alunos de duas turmas diferentes
ap√≥s um novo m√©todo de ensino. Realize o teste de hip√≥teses para comparar as m√©dias das duas turmas usando um n√≠vel de signific√¢ncia de 0,05.
As notas das duas turmas s√£o:

```python
#Notas dos alunos das duas turmas
Turma_A = [85, 78, 90, 88, 76, 95, 89, 84]
Turma_B = [82, 75, 85, 80, 79, 88, 83, 77]

#Teste de hipoteses
#H0 - as m√©dias s√£o iguais
#H1 - as m√©dias s√£o diferentes

from scipy import stats

t_estatistico, p_valor = stats.ttest_ind(Turma_A, Turma_B)
print(f"Estat√≠stica t √©: {t_estatistico:.1f}")
print(f"Valor p √©: {p_valor:.1f}")

if p_valor < 0.05:
  print("Rejeitamos a hipotese nula, as m√©dias s√£o diferentes.")
else:
  print("Aceitamos a hipotese nula, porque as m√©dias s√£o iguais.")
```

## üß™ Atividade Estatistica - Teste de Hipoteses

Esse teste √© um procedimento usado para avaliar se uma afirma√ß√£o sobre uma popula√ß√£o pode ser rejeitada ou n√£o, com base em dados de uma amostra.

Ele envolve a formula√ß√£o de duas hip√≥teses (nula e alternativa), a coleta de dados e o uso de uma estat√≠stica de teste para determinar se h√° evid√™ncias suficientes para rejeitar a hip√≥tese nula. Este processo √© crucial para tomar decis√µes informadas,

3. Uma empresa testou a efic√°cia de um novo software em dois per√≠odos de tempo. Os tempos (em minutos)
para concluir uma tarefa antes e depois do uso do software s√£o registrados abaixo. Determine se houve
uma melhoria significativa no tempo de conclus√£o da tarefa.
Antes: [30, 28, 35, 33, 40, 29, 32, 36] Depois: [25, 22, 30, 28, 35, 27, 30, 31]

```python
#H0 - n√£o houve melhora de tempo
#H1 - houve melhora significativa de tempo

from scipy import stats

antes = [30, 28, 35, 33, 40, 29, 32, 36]
depois = [25, 22, 30, 28, 35, 27, 30, 31]

t_estatistico, p_valor = stats.ttest_rel(antes, depois)
print(f"Estat√≠stica t √©: {t_estatistico:.1f}")
print(f"Valor p √©: {p_valor:.1f}")

if p_valor < 0.05:
  print("Rejeitamos a hipotese nula, pois, houve melhora significativa de tempo.")
else:
  print("Aceitamos a hipotese nula, pois, n√£o houve melhora significativa de tempo.")
```

## üß™ Atividade Estatistica - Teste Quiquadrado

Essa ferramenta √© usada para analisar dados categ√≥ricos, verificando se h√° uma associa√ß√£o significativa entre duas ou mais vari√°veis. Ele compara as frequ√™ncias observadas em categorias com as frequ√™ncias esperadas sob uma hip√≥tese nula.

4. Uma pesquisa foi realizada para verificar a prefer√™ncia de 100 pessoas por dois tipos de bebidas, A e B, em duas cidades diferentes, X e Y. A tabela de conting√™ncia a seguir mostra os resultados:

    * Cidade X Cidade Y Bebida A 30 40, Bebida B 20 10.

Verifique se h√° uma rela√ß√£o significativa entre a cidade e a prefer√™ncia pela bebida.

```python
#H0 - a prefer√™ncia pela bebida √© independente da cidade.
#H1 - a prefer√™ncia pela bebida depende da cidade.

from scipy.stats import chi2_contingency

dados = [[30, 20], [40, 10]]
estatistica_chi2, p_valor, graus_de_liberdade, tabela_de_frequencia = chi2_contingency(dados)

print(f"Estat√≠stica Qui-Quadrado √©: {estatistica_chi2: .2f}")
print(f"P valor √©: {p_valor:.2f}")

if p_valor < 0.05:
  print("Rejeitamos a hipotese nula, a prefer√™ncia pela bebida depende da cidade.")
else:
  print("Aceitamos a hipotese nula, a prefer√™ncia pela bebida √© independente da cidade.")
```

## üß™ Atividade Estatistica - Teste ANOVA

5. Um pesquisador deseja testar se tr√™s diferentes tipos de fertilizantes t√™m um efeito significativo no crescimento de plantas.
Os crescimentos das plantas (em cm) para cada fertilizante s√£o dados abaixo:

    * Fertilizante A: [20, 22, 19, 21, 20]
    * Fertilizante B: [18, 20, 17, 19, 18]
    * Fertilizante C: [25, 27, 26, 28, 26]

  Realize uma ANOVA de uma via para verificar se h√° uma diferen√ßa significativa entre os grupos.

```python
#H0 - as m√©dias dos fertilizantes s√£o iguais
#H1 - pelo menos uma das m√©dias √© diferente

import scipy.stats as stats

dados = [[20, 22, 19, 21, 20], [18, 20, 17, 19, 18], [25, 27, 26, 28, 26]]

anova_resultado = stats.f_oneway(*dados)

print(f"Estat√≠stica ANOVA √©: {anova_resultado.statistic:.2f}")
print(f"P_valor √©: {anova_resultado.pvalue:.2f}")

if anova_resultado.pvalue < 0.05:
  print("Rejeitamos a hipotese nula, porque pelo menos uma das m√©dias √© diferente.")
```
